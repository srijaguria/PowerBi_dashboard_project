```python
import torch
from transformers import AutoTokenizer
```

```python
tokenizer=AutoTokenizer.from_pretrained('bert-base-cased')
```

```python
inputs="Rahul loves to play cricket"
```

```python
tokens=tokenizer.encode_plus(inputs)
```

```python
tokens
```

```python
tokenizer.decode(tokens.input_ids)
```
